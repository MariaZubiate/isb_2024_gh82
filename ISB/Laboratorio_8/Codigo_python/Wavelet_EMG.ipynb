#LOADING DATA

```python
import numpy as np
import matplotlib.pyplot as plt
archivos = ['pulgar_reposo.txt', 'pulgar_flexion.txt', 'pulgar_contrafuerza.txt']
for archivo in archivos:
  # Cargar los datos de la seal EEG desde el archivo de texto
  EMG_data = np.loadtxt(archivo)
  # Seleccionar la ltima columna que contiene los valores de la seal EEG
  columna_EMG = EMG_data[:, -1]
  """Debido a que la seal muestreada por el EEG del BiTalino son del ADC de 10 bits; entonces esta data oscila
  de 0-1023. Pero considerando que el voltaje de operacin es de 3.3V, se debe hacer una conversin de 0-1023 a 0-3.3V"""
  v = ((columna_EMG-507)/1023)*3.3*1000
  """Dado que el BiTalino te brinda el total de datos y no el tiempo, podemos obtener
  la duracin del muestreo gracias a la frecuencia de meustreo del BiTalino: 1000 Hz"""
  fs = 1000
  N = len(v)
  """Teniendo la duracin y la cantidad de muestreo, se realiz otra lista para el eje X
  que equivale al Tiempo, con intervalos iguales para obtener un tamao de lista igual al del Eje Y"""
  t = np.linspace(0, N/fs, N)
  #Plotear la seal EEG en el dominio del tiempo
  plt.plot(t, v)
  plt.xlabel('Tiempo (s)')
  plt.ylabel('Voltaje (mV)')
  plt.title('Seal EEG - '+archivo.replace('_', ' ').replace('.txt', '').title())
  plt.grid()
  plt.show()
  #FFT
  f = np.linspace(-fs/2, fs/2, N)
  V = np.fft.fftshift(np.fft.fft(v))
  plt.plot(f, abs(V))
  plt.xlabel('Frecuencia (Hz)')
  plt.ylabel('Magnitud')
  plt.title('FFT de la Seal EMG - '+archivo.replace('_', ' ').replace('.txt', '').title())
  plt.show()
```

#Filtro Wavelet

```python
import numpy as np
import matplotlib.pyplot as plt
import pywt

# Funcin para realizar y plotear el filtro wavelet
def aplicar_filtro_wavelet(archivo):
    # Cargar los datos de la seal EEG desde el archivo de texto
    EMG_data = np.loadtxt(archivo)
    # Seleccionar la ltima columna que contiene los valores de la seal EEG
    columna_EMG = EMG_data[:, -1]
    # Convertir los datos del rango 0-1023 a voltaje en mV
    v = ((columna_EMG - 507) / 1023) * 3.3 * 1000
    # Parmetros de la seal
    fs = 1000
    N = len(v)
    t = np.linspace(0, N/fs, N)

    # Aplicar la Transformada Wavelet
    wavelet = 'sym8'
    level = 9
    coeffs = pywt.wavedec(v, wavelet, level=level)

    # Eliminar coeficientes especficos D1, D8, D9 y A9
    coeffs[0] = np.zeros_like(coeffs[0])  # A9
    coeffs[1] = np.zeros_like(coeffs[1])  # D9
    coeffs[2] = np.zeros_like(coeffs[2])  # D8
    coeffs[-1] = np.zeros_like(coeffs[-1])  # D1

    # Reconstruccin de la seal filtrada
    v_filtrada = pywt.waverec(coeffs, wavelet)

    # Plotear la seal original y filtrada en el dominio del tiempo
    plt.figure(figsize=(12, 6))
    plt.subplot(2, 1, 1)
    plt.plot(t, v, label='Seal Original', linewidth=0.5)
    plt.xlabel('Tiempo (s)')
    plt.ylabel('Voltaje (mV)')
    plt.title('Seal EMG Original - ' + archivo.replace('_', ' ').replace('.txt', '').title())
    plt.legend()
    plt.margins(0,0.2)
    plt.xlim([1,12])
    #plt.grid()

    plt.subplot(2, 1, 2)
    plt.plot(t, v_filtrada, linewidth=0.5, label='Seal Filtrada')
    plt.xlabel('Tiempo (s)')
    plt.ylabel('Voltaje (mV)')
    plt.title('Seal EMG Filtrada - ' + archivo.replace('_', ' ').replace('.txt', '').title())
    plt.legend()
    plt.margins(0,0.2)
    plt.xlim([1,12])
    plt.tight_layout()
    plt.show()
    #plt.grid()

    # FFT de la seal original y filtrada
    f = np.linspace(-fs/2, fs/2, N)
    V = np.fft.fftshift(np.fft.fft(v))
    V_filtrada = np.fft.fftshift(np.fft.fft(v_filtrada))

    plt.figure(figsize=(12, 6))
    plt.subplot(2, 1, 1)
    plt.plot(f, np.abs(V),  linewidth=0.5, label='FFT Seal Original')
    plt.xlabel('Frecuencia (Hz)')
    plt.ylabel('Magnitud')
    plt.title('FFT de la Seal EMG Original - ' + archivo.replace('_', ' ').replace('.txt', '').title())
    plt.legend()
    plt.margins(0,0)
    #plt.grid()

    plt.subplot(2, 1, 2)
    plt.plot(f, np.abs(V_filtrada), linewidth=0.5, label='FFT Seal Filtrada')
    plt.xlabel('Frecuencia (Hz)')
    plt.ylabel('Magnitud')
    plt.title('FFT de la Seal EMG Filtrada - ' + archivo.replace('_', ' ').replace('.txt', '').title())
    plt.legend()
    plt.margins(0,0)
    plt.tight_layout()
    plt.show()
    #plt.grid()

# Lista de archivos
archivos = ['pulgar_reposo.txt', 'pulgar_flexion.txt', 'pulgar_contrafuerza.txt']

# Aplicar el filtro wavelet a cada archivo
for archivo in archivos:
    aplicar_filtro_wavelet(archivo)
```
