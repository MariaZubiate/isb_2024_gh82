#deep_breath

```python
import pandas as pd
import numpy as np
import os
import requests

# Cargar los datos desde el archivo de texto
data_cruda = np.loadtxt('/content/deep_breath.txt')

# Extraer la última columna y convertirla a milivoltios
data_cruda = data_cruda[:, -1]
signal = ((data_cruda - 507) / 1023) * 3.3 * 1000

# Definir la frecuencia de muestreo
fs = 1000  # Frecuencia de muestreo en Hz

# Seleccionar los primeros 5 segundos de datos
signal_data = list(signal[:-1 * fs])

# Generar los datos de tiempo correspondientes
time_data = np.linspace(0, len(signal_data) / fs, len(signal_data))

# Convertir el tiempo a milisegundos
time_data_ms = time_data * 1000

# Crear un DataFrame con los datos de tiempo y señal
df = pd.DataFrame({'timestamp': time_data_ms, 'EMG': signal_data})

# Definir la longitud de la ventana en segundos
window_length_seconds = 1  # Cambiar este valor según sea necesario

# Calcular el tamaño de la ventana en muestras
window_size = window_length_seconds * fs

# Lista para almacenar los nombres de los archivos generados
files = []

# Dividir los datos en ventanas y ajustar los tiempos de cada ventana
num_windows = len(df) // window_size
for i in range(num_windows):
    start_time = i * window_size / fs * 1000  # Convertir el tiempo de inicio a milisegundos
    window_data = df.iloc[i * window_size:(i + 1) * window_size].copy()
    window_data['timestamp'] = np.linspace(start_time, start_time + window_size / fs * 1000, window_size, endpoint=False)
    output_file = f'ecg_deep_breath._{i + 1}.csv'
    window_data.to_csv(output_file, index=False)
    files.append(output_file)
    print(f'Archivo CSV guardado como {output_file}')

# Verificar si quedan datos que no alcanzan a completar una ventana
if len(df) % window_size != 0:
    start_time = num_windows * window_size / fs * 1000
    window_data = df.iloc[num_windows * window_size:].copy()
    window_data['timestamp'] = np.linspace(start_time, start_time + len(window_data) / fs * 1000, len(window_data), endpoint=False)
    output_file = f'ecg_deep_breath._{num_windows + 1}.csv'
    window_data.to_csv(output_file, index=False)
    files.append(output_file)
    print(f'Archivo CSV guardado como {output_file}')

# Imprimir la lista de archivos generados
print('Archivos generados:', files)
```
