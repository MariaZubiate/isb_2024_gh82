```python
import numpy as np
import matplotlib.pyplot as plt

# Cargar los datos de la seal EEG desde el archivo de texto

#Seal base
EEG_data_1 = np.loadtxt('En Reposo.txt')

#Ojos cerrados 30 segundos
EEG_data_2 = np.loadtxt('Durante el Parpadeo.txt')

#Abrir y cerrar ojos
EEG_data_3 = np.loadtxt('Despus del Parpadeo.txt')

#Preguntas matemticas
EEG_data_4 = np.loadtxt('Resolver Problemas Mentalmente.txt')

# Seleccionar la ltima columna que contiene los valores de la seal EEG
for x in [EEG_data_1, EEG_data_2, EEG_data_3, EEG_data_4]:
    columna_EEG = x[:, -1]
    """Debido a que la seal muestreada por el EEG del BiTalino son del ADC de 10 bits; entonces esta data oscila
    de 0-1023. Pero considerando que el voltaje de operacin es de 3.3V, se debe hacer una conversin de 0-1023 a 0-3.3V"""
    columna_Volt = ((columna_EEG-507)/1023)*3.3*1000
    """Dado que el BiTalino te brinda el total de datos y no el tiempo, podemos obtener
    la duracin del muestreo gracias a la frecuencia de meustreo del BiTalino: 1000 Hz"""
    fs = 1000
    total_data = len(columna_Volt)
    time_total = total_data/fs
    """Teniendo la duracin y la cantidad de muestreo, se realiz otra lista para el eje X
    que equivale al Tiempo, con intervalos iguales para obtener un tamao de lista igual al del Eje Y"""
    time_data = np.linspace(0, time_total, total_data)
    #Para visualizar solo una porcin del grfico
    t_start = 8.5 # @param {type:"number"}
    t_end = 9.2  # @param {type:"number"}
    index_start = round(t_start*(total_data-1)/time_total)
    index_end = round(t_end*(total_data-1)/time_total)

    #Plotear la seal EEG en el dominio del tiempo
    plt.plot(time_data, columna_Volt)
    plt.xlabel('Tiempo (s)')
    plt.ylabel('Voltaje (mV)')
    if x is EEG_data_1:
        plt.title('Seal EEG en Reposo')
    elif x is EEG_data_2:
        plt.title('Seal EEG durante el Parpadeo')
    elif x is EEG_data_3:
        plt.title('Seal EEG despus del Parpadeo')
    else:
        plt.title('Seal EEG durante la Resolucin de Problemas Mentalmente')
    plt.grid()
    plt.show()

    #Ploteo de la seal de EEG en un intervalo
    plt.plot(time_data, columna_Volt)
    plt.xlabel('Tiempo (s)')
    plt.ylabel('Voltaje (mV)')
    if x is EEG_data_1:
        plt.title('Intervalo de la Seal EEG en Reposo')
    elif x is EEG_data_2:
        plt.title('Intervalo de la Seal EEG durante el Parpadeo')
    elif x is EEG_data_3:
        plt.title('Intervalo de la Seal EEG despus del Parpadeo')
    else:
        plt.title('Intervalo de la Seal EEG durante la Resolucin de Problemas Mentalmente')
    plt.xlim([10, 20])
    plt.grid()
    plt.show()
```
